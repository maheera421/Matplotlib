# Histogram

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = [1, 3, 3, 3, 4,7,9, 9, 7, 5, 6]
```

data is computed for graph as in stats

```python
plt.hist(data)
plt.show()
```

Changing aesthetics

```python
plt.hist(data, color = "g", bins = 4)
plt.show()
```

bins/bars is reduced

Now working with a dataset in iris.csv file

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.hist(iris['petal_length'], bins = 50, color='g')
plt.show()
```
