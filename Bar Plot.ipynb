## Bar Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
students = {"Maheera": 72, "Ayesha": 87, "Mairal": 55}
```

```python
names = list(students.keys())
values = list(students.values())
```

```python
plt.bar(names, values)
plt.title("Vertical Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names, values, color = 'g')
plt.title("Horizontal Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
