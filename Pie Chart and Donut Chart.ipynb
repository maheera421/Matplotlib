# Pie Chart

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fruit = ['Apple', 'Mango', 'Orange', 'Guava']
quantity = [40, 50, 10, 30]
```

```python
plt.pie(quantity, labels=fruit)
plt.title("Simple pie chart")
plt.show()
```

draws the chart with relative average.

```python
plt.pie(quantity, labels=fruit, autopct = '%0.1f%%', colors=['red','yellow', 'orange', 'green'])
plt.title("Enhanced pie chart.")
plt.show()
```

autopct is used to set the decimal points of the percentage and colors are set using list of colors respectively

# Donut Chart

```python
plt.pie(quantity, labels=fruit, radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.show()
```

basically it has two pie charts (circles)

2nd pie chart (inner one) is of white color
