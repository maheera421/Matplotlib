```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

Note that you wrote plt.plot and plt.show again to output the graph 

```python
plt.plot(x,y, color="green", linestyle=":", linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
y1 = x*2
y2 = x*3
plt.plot(x,y1, color="green", linestyle=":", linewidth=3)
plt.plot(x,y2, color="red", linestyle="-.", linewidth=2)
plt.title("Line Plot with two lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

no need to initialize x again since x remains same y value changes for different lines

```python
plt.subplot(1,2,1)
plt.plot(x,y1, color="green", linewidth=3)
plt.title("First Subplot column-wise")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(1,2,2)
plt.plot(x,y2, color="red", linewidth=3)
plt.title("Second Subplot column-wise")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

no need to initialize x y1 and y2 variables again since we already have the data ready

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color="green", linewidth=3)
plt.title("First Subplot row-wise")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2,1,2)
plt.plot(x,y2, color="red", linewidth=3)
plt.title("Second Subplot row-wise")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
