# Scatter Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
y = [5, 8, 3, 9, 6, 2, 8, 5, 7]
```

```python
plt.scatter(x, y)
plt.title("Simple scatter plot")
plt.show()
```

```python
plt.scatter(x, y, marker="*", c='g', s=200)
plt.title("Advanced scatter plot")
plt.show()
```

s represents size of marker

```python
y1 = [5, 8, 3, 9, 6, 2, 8, 5, 7]
y2 = [6, 9, 5, 7, 4, 4, 6, 9, 5]
plt.scatter(x, y1, marker = "*", c='g', s=100)
plt.scatter(x, y2, marker = ".", c='r', s=200)
plt.title("Scatter plot for two y values")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x, y1, marker = "*", c='g', s=100)


plt.subplot(1,2,2)
plt.scatter(x, y2, marker = ".", c='r', s=200)
plt.show()
```

Note that in subplots we use only one plt.show() to show both plots
